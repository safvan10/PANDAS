```python
import pandas as pd
df1=pd.DataFrame([[1,2,3],[4,5,6]])
df2=pd.DataFrame([[7,8,9],[10,11,12]])

def myfunc(a,b):
    if(a.sum()>b.sum()):
        return a
    else:
        return b
    
print(df1.combine(df2,myfunc))
```
